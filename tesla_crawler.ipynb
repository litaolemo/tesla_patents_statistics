```python
print(len(res_patent_list))

headers = {
'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
'accept-encoding': 'gzip, deflate, br',
'accept-language': 'zh-CN,zh;q=0.9',
'cache-control': 'max-age=0',
'cookie': '',
'dnt': '1',
'sec-ch-ua': '"Chromium";v="112", "Google Chrome";v="112", "Not:A-Brand";v="99"',
'sec-ch-ua-mobile': '?0',
'sec-ch-ua-platform': '"Windows"',
'sec-fetch-dest': 'document',
'sec-fetch-mode': 'navigate',
'sec-fetch-site': 'none',
'sec-fetch-user': '?1',
'upgrade-insecure-requests': '1',
'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36',

}
from bs4 import BeautifulSoup


def parse_data(res_data_text,patent_Citations,key):
    bf = BeautifulSoup(res_data_text,'html.parser')
    res_div = bf.find('div',class_='tr style-scope patent-result')
    soup = BeautifulSoup(res_data_text,'html.parser')
    res = soup.find_all("tr")
    n = 0
    for r in res:
        if r.attrs:
            if r.attrs.get("itemprop") in ["backwardReferencesOrig", "forwardReferences", "backwardReferencesFamily","backwardReferences","forwardReferencesFamily"]:
                p_num = r.text.split("\n")[3]
                patent_Citations[key][p_num] = {}
                if "backward" in r.attrs.get("itemprop"):
                    cite_type = "backward"
                else:
                    cite_type = "forward"
                patent_Citations[key][p_num]["cite_type"] = cite_type
                if r.attrs.get("itemprop") in ("backwardReferencesFamily","forwardReferencesFamily"):
                    patent_Citations[key][p_num]["Priority date"] = r.text.split("\n")[8]
                    patent_Citations[key][p_num]["Publication date"] = r.text.split("\n")[9]
                    patent_Citations[key][p_num]["Assignee"] = r.text.split("\n")[10]
                    if r.text.split("\n")[11]:
                        patent_Citations[key][p_num]["Title"] = r.text.split("\n")[11]
                    else:
                        try:
                            patent_Citations[key][p_num]["Title"] = r.text.split("\n")[12]
                        except:
                            patent_Citations[key][p_num]["Title"] = ""

                elif  r.attrs.get("itemprop") in ("backwardReferences","backwardReferencesOrig"):
                    patent_Citations[key][p_num]["Priority date"] = r.text.split("\n")[7]
                    patent_Citations[key][p_num]["Publication date"] = r.text.split("\n")[8]
                    patent_Citations[key][p_num]["Assignee"] = r.text.split("\n")[9]
                    patent_Citations[key][p_num]["Title"] = r.text.split("\n")[10]
                elif "forwardReferences" == r.attrs.get("itemprop"):
                    patent_Citations[key][p_num]["Priority date"] = r.text.split("\n")[8]
                    patent_Citations[key][p_num]["Publication date"] = r.text.split("\n")[9]
                    patent_Citations[key][p_num]["Assignee"] = r.text.split("\n")[10]
                    if r.text.split("\n")[11]:
                        patent_Citations[key][p_num]["Title"] = r.text.split("\n")[11]
                    else:

                        patent_Citations[key][p_num]["Title"] = r.text.split("\n")[12]

                if patent_Citations[key][p_num]["Title"] == "":
                    print("error",r.text.split("\n"),r.attrs.get("itemprop"))
                n += 1
    print("total ",n," citations")
    return patent_Citations

```
